```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Function to show multiple images side by side
def show_images(images, titles, cmap='gray'):
    plt.figure(figsize=(10, 5))
    for i in range(len(images)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(images[i], cmap=cmap)
        plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
img = cv2.imread('/content/drive/MyDrive/Classroom/CSE438: Digital Image Processing Lab 8A/Practice/animal.png', 0)

# Show the original image
show_images([img], ['Original Image'])
```

```python
# Resize down
small = cv2.resize(img, (0, 0), fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)

# Resize back up to original for display
upscaled = cv2.resize(small, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)

show_images([img, upscaled], ["Original", "Transformation 1"])
```

```python
level = 2
quantized = np.floor(img / (256 / level)) * (256 / level)
quantized = quantized.astype(np.uint8)

show_images([img, upscaled, quantized], ['Original Image', "Transformation 1", "Transformation 2"])
```
